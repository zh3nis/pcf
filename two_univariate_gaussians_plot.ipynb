# Two Univariate Gaussian PDFs

Plot $\mathcal{N}(+\mu, 1)$ in red and $\mathcal{N}(-\mu, 1)$ in blue.

```python
import numpy as np
import matplotlib.pyplot as plt

mu = 2.0  # Change this value as needed
sigma2 = 1.0
output_path = 'two_univariate_gaussians_plot.png'

plt.rcParams.update({
    'font.size': 16,
    'axes.titlesize': 20,
    'axes.labelsize': 18,
    'xtick.labelsize': 16,
    'ytick.labelsize': 16,
    'legend.fontsize': 16,
})

def gaussian_pdf(x, mean, var=1.0):
    return np.exp(-0.5 * ((x - mean) ** 2) / var) / np.sqrt(2 * np.pi * var)

x = np.linspace(-6, 6, 1000)
pdf_plus = gaussian_pdf(x, mean=mu, var=sigma2)
pdf_minus = gaussian_pdf(x, mean=-mu, var=sigma2)

plt.figure(figsize=(10, 6))
plt.plot(x, pdf_plus, color='red', linewidth=2.5, label=rf'$\mathcal{{N}}(+\mu, 1)$')
plt.plot(x, pdf_minus, color='blue', linewidth=2.5, label=rf'$\mathcal{{N}}(-\mu, 1)$')
plt.axvline(mu, color='red', linestyle='--', alpha=0.4)
plt.axvline(-mu, color='blue', linestyle='--', alpha=0.4)
#plt.title('Two Univariate Gaussian PDFs with Unit Variance')
plt.xlabel('x')
plt.ylabel('Density')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.12), ncol=2, frameon=False)
plt.grid(alpha=0.3)
plt.subplots_adjust(bottom=0.24)
plt.tight_layout()
plt.savefig(output_path, dpi=600, bbox_inches='tight')
plt.show()

print(f'Saved figure to: {output_path}')

```
